```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_discrete_pmf(x_values, probabilities):
    # Create a dictionary to map x_values to probabilities
    data = {x: p for x, p in zip(x_values, probabilities)}

    # Plot the PMF using histplot
    sns.histplot(data=data, discrete=True, stat="probability", linewidth=0, kde=False)

    # Set labels and title
    plt.xlabel('X')
    plt.ylabel('Probability')
    plt.title('Discrete Probability Distribution PMF')

    # Show the plot
    plt.show()

# Example data
x_values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.4, 0.15, 0.15]

# Call the plotting function
plot_discrete_pmf(x_values, probabilities)
```
